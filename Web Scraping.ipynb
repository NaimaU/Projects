Scraping from HTML to Pandas Dataframe + CSV

```python
''' 
Using the Regex wikipedia page.
https://en.wikipedia.org/wiki/Regular_expression
'''
```

```python
url = 'https://en.wikipedia.org/wiki/Regular_expression'

# headers are necessary to prevent 403 error
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0 Safari/537.36"
}
response = requests.get(url, headers=headers)
print(response)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup)
```

```python
table = soup.find('table', 'wikitable sortable')
```

```python
# extract headers

tbl_columns = table.find_all('th')
print(tbl_columns)
```

```python
tbl_columns = [item.text.strip() for item in tbl_columns]
print(tbl_columns)
```

```python
# headers into dataframe

import pandas as pd
df = pd.DataFrame(columns = tbl_columns)
```

```python
df
```

```python
info = table.find_all('tr')[1:]
print(info)
```

```python
for row in info:
    row_data = row.find_all('td')
    # print(row_data)

    # extract text from  each td tag
    indiv_row_data = [element.text.strip() for element in row_data]
    # print(indiv_row_data)

    # add each row to existing pandas dataframe
    length = len(df)
    df.loc[length] = indiv_row_data
```
