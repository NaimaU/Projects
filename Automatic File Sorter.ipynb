```python
import os, shutil, magic
```

```python
path = r"C:\Users\naima\Documents\Portfolio Projects\\"
```

```python
files = os.listdir(path)
print(files)
```

```python
# extract file type via string
# str.endswith()
# assuming all file names end with '.file_type'
for file in files:
    file_type = file.rpartition('.')
    print(file_type)

    file_type = file_type[2]
    print(file_type)
```

```python
# extract file type using stat
with os.scandir(path) as dir_entries:
    for entry in dir_entries:
        info = entry.stat()
        print(info)    # can't actually access filetype...can sep files with other attributes instead
```

```python
# extract file type using magic library
for file in files:
    file_type = magic.from_file(path + file)
    print(file_type)
```

```python
# extract file type using filetype library
import filetype

for file in files:
    file_type = filetype.guess(path + file)
    print(file_type)
```

```python
# will be using the first method (extract via string) to make directories and sort
```

```python
# create directory if it doesn't exist. then move file to correct directory

for file in files:
    file_type = file.rpartition('.')[2]

    # create directory if it doesn't exist
    if not(os.path.exists(path + file_type)):
        os.mkdir(path + file_type)

    # move file to correct directory. src -> dst
    shutil.move(path + file, path + file_type)
    
```
